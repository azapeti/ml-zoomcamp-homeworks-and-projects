## Question 3

Now let's experiment with the `n_estimators` parameter

* Try different values of this parameter from 10 to 200 with step 10.
* Set `random_state` to `1`.
* Evaluate the model on the validation dataset.


After which value of `n_estimators` does RMSE stop improving?
Consider 3 decimal places for calculating the answer.

- 10
- 25
- 80
- 200

````python
scores = []
for e in range(10, 201, 10):
    rf = RandomForestRegressor(n_estimators=e, random_state=1, n_jobs=-1)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_val)
    rmse = root_mean_squared_error(y_val, y_pred)
    scores.append((e, rmse))
    
````

````python
rf_scores = pd.DataFrame(scores, columns=['n_estimators', 'rmse'])
rf_scores.rmse = rf_scores.rmse.round(3)
rf_scores.sort_values(by='rmse')
````

````python
plt.plot(rf_scores.n_estimators, rf_scores.rmse)
````

## Question 4

Let's select the best `max_depth`:

* Try different values of `max_depth`: `[10, 15, 20, 25]`
* For each of these values,
  * try different values of `n_estimators` from 10 till 200 (with step 10)
  * calculate the mean RMSE 
* Fix the random seed: `random_state=1`


What's the best `max_depth`, using the mean RMSE?

* 10
* 15
* 20
* 25

````python
scores = []
for d in [10, 15, 20, 25]:
    for e in range(10, 201, 10):
        rf = RandomForestRegressor(max_depth=d,
                                   n_estimators=e,
                                   random_state=1,
                                   n_jobs=-1)
        rf.fit(X_train, y_train)
        y_pred = rf.predict(X_val)
        rmse = root_mean_squared_error(y_val, y_pred)
        scores.append((d, e, rmse))
````

````python
rf_scores = pd.DataFrame(scores, columns=['max_depth', 'n_estimators', 'rmse'])
mean_RMSE_by_max_depth = rf_scores.groupby('max_depth')['rmse'].mean()
mean_RMSE_by_max_depth.sort_values()
````

````python
rf_scores
````

````python
df = rf_scores
# Plotting
fig, axes = plt.subplots(2, 2, figsize=(10, 8), sharex=True, sharey=True)
fig.suptitle("RMSE vs. n_estimators for Different max_depth Values")

# Iterate over unique max_depth values and plot each in a subplot
for ax, (max_depth, group) in zip(axes.flat, df.groupby('max_depth')):
    ax.plot(group['n_estimators'], group['rmse'], marker='o')
    ax.set_title(f'max_depth = {max_depth}')
    ax.set_xlabel("n_estimators")
    ax.set_ylabel("RMSE")
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit suptitle
plt.show()
````

# Question 5

We can extract feature importance information from tree-based models. 

At each step of the decision tree learning algorithm, it finds the best split. 
When doing it, we can calculate "gain" - the reduction in impurity before and after the split. 
This gain is quite useful in understanding what are the important features for tree-based models.

In Scikit-Learn, tree-based models contain this information in the
[`feature_importances_`](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html#sklearn.ensemble.RandomForestRegressor.feature_importances_)
field. 

For this homework question, we'll find the most important feature:

* Train the model with these parameters:
  * `n_estimators=10`,
  * `max_depth=20`,
  * `random_state=1`,
  * `n_jobs=-1` (optional)
* Get the feature importance information from this model


What's the most important feature (among these 4)? 

* `study_hours_per_week`
* `attendance_rate`
* `distance_to_school`
* `teacher_quality`



````python
rf = RandomForestRegressor(max_depth=20,
                           n_estimators=10,
                           random_state=1,
                           n_jobs=-1)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)
rmse = root_mean_squared_error(y_val, y_pred)
scores.append((d, e, rmse))
````

````python
feature_names = list(dv.get_feature_names_out(X_train))
````

````python
feature_importances = rf.feature_importances_
feature_importances
````

````python
importances_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

print(importances_df)

plt.figure(figsize=(10, 6))
plt.barh(importances_df['Feature'], importances_df['Importance'], color='skyblue')
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.title("Feature Importances in RandomForestRegressor")
plt.gca().invert_yaxis()
plt.show()
````

````python
def parse_xgb_output(output):
    results = []

    for line in output.stdout.strip().split('\n'):
        it_line, train_line, val_line = line.split('\t')

        it = int(it_line.strip('[]'))
        train = float(train_line.split(':')[1])
        val = float(val_line.split(':')[1])

        results.append((it, train, val))
    
    columns = ['num_iter', 'train_rmse', 'val_rmse']
    df_results = pd.DataFrame(results, columns=columns)
    return df_results
````
